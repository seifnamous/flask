```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import seaborn as sns
import cv2
```

```python
path = '/home/namous/ai/simplon/manel/flask_img/static/assets/113.jpg'
img_loaded = cv2.imread(path, cv2.IMREAD_UNCHANGED)
img_loaded.shape
```

```python
print(img_loaded)
```

```python
scale_percent = 60 
width = int(img_loaded.shape[1] * scale_percent / 100)
height = int(img_loaded.shape[0] * scale_percent / 100)
dim = (width, height)
img_resized = cv2.resize(img_loaded, dim, interpolation = cv2.INTER_AREA)
img_resized.shape
```

```python
F1 = 1/ 100 * np.array([ [1, 1, 1], [1, 1, 1], [1, 1, 1]])
img_filtered=cv2.filter2D(img_resized,-1,kernel=F1)
```
